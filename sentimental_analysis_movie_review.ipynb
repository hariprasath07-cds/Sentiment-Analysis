# TEXT PREPROCESSING WITH NLTK

Preprocessing steps:

Convert to lowercase
Remove punctuation and numbers
Tokenize (split into words)
Remove stop words
Lemmatize (reduce to root form)

```python
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

def preprocess(text):
    # Convert to lowercase
    text = text.lower()

    # Remove punctuation and numbers
    text = re.sub(r'[^a-z\s]', '', text)

    # Tokenize
    tokens = word_tokenize(text)

    # Remove stop words
    tokens = [token for token in tokens if token not in stop_words]

    # Lemmatize: convert it back to root word
    tokens  = [lemmatizer.lemmatize(token) for token in tokens]

    # Join tokens back to string
    return " ".join(tokens)
```

```python
# build the full dataset as list of text and labels

text = []
label = []

for category in movie_reviews.categories():
    for file_id in movie_reviews.fileids(category):
        raw_text = movie_reviews.raw(file_id)
        text.append(raw_text)
        label.append(category)
```

# FEATURE EXTRACTION WITH TF- IDF


```python
pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(cleaned_text,
                                                    label,
                                                    test_size=0.2,
                                                    random_state=42,
                                                    stratify=label)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
len(y_train)
```

```python
len(y_test)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=10000) # 10,000 most important words
```

```python
X_train_tfidf=vectorizer.fit_transform(X_train)
```

```python
X_test_tfidf = vectorizer.transform(X_test)
```

```python
# Some of the vocabulary learned by the vectorizer
len(vectorizer.get_feature_names_out())
```

## TRAIN AND COMPARE CLASSIFIERS

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
```

```python
from sklearn.metrics import accuracy_score,ConfusionMatrixDisplay,classification_report
```

```python
nb_model = MultinomialNB()
nb_model.fit(X_train_tfidf, y_train)
```

```python
y_pred_nb = nb_model.predict(X_test_tfidf)
accuracy_nb = accuracy_score(y_test, y_pred_nb)
```

```python
accuracy_nb
```

```python
lr_model = LogisticRegression()
lr_model.fit(X_train_tfidf, y_train)
```

```python
y_pred_lr = lr_model.predict(X_test_tfidf)
accuracy_lr = accuracy_score(y_test, y_pred_lr)
```

```python
accuracy_lr
```

```python
print(classification_report(y_test, y_pred_lr))
```

```python
ConfusionMatrixDisplay.from_estimator(lr_model, X_test_tfidf, y_test)
```

```python
lr_model.coef_[0]
```
